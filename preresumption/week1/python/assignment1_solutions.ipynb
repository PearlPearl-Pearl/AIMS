### Question 8
A year is a leap year if it is divisible by 4, except that years divisible by 100 are not leap years unless they are also divisible by 400. Write a program that asks the user for a year and prints out whether it is a leap year or not. For example, the years 1700, 1800, and 1900 are not leap years, but the years 1600, 1996, and 2000 are.

````python
year = int(input('Enter the year: '))

if year%4==0:
  if year%100==0:
    if year%400==0:
      print('The year is a leap year.')
    else:
      print('Not a leap year.')
  else:
    print('The year is a leap year.')
else:
  print('Not a leap year')
````

### Question 10
The Fibonacci numbers form a sequence that has the following generating function
$$ F_0 = 1,\quad F_1 = 1$$
$$F_i = F_{i-1} + F_{i-2} \quad , \quad i = 2,3, \dots $$

By considering the terms in the Fibonacci sequence whose values do not exceed four million, find the sum of the even-valued terms.

````python
previous_previous_term = 1
previous_term = 1

current_term = previous_previous_term + previous_term

even_terms_less_4 = []

while current_term < 4000000:
  if current_term%2==0:
    even_terms_less_4.append(current_term)
  previous_previous_term = previous_term
  previous_term = current_term
  current_term = previous_previous_term + previous_term


print(f'The sum of the even-valued terms Fibonacci sequence whose value does not exceed four million is {sum(even_terms_less_4)}')
````

### Question 11
Ask the user to enter a number, and print the squares of the first five multiples of the number, separated by >>>
```
Example

Input: 3  
Output: 9>>>36>>>81>>>144>>>225
```

````python
num = int(input('Enter a number: '))

for i in range(num, (5*num)+1, num):
  if i==num*5:
    end = ''
  else:
    end='>>>'
  print(f'{i**2}', end=end)
````

### Question 12
Write a Python program that checks if a number is prime. Prompt the user to enter a number and return whether it is prime or not.

````python
num = int(input('Enter a positive integer '))

factors = []

while num<0 or not isinstance(num, int):
  num = int(input('You need to enter an integer which is greater than 0 '))
if num == 0 or num == 1:
   print('Number is not prime')
else:
   for i in range(2, num):
    if num%i==0:
        factors.append(i)
   if len(factors)==0:
      print(f'Number {num} is prime')
   else:
      print(f'Number {num} is not prime and the extra factors are: {factors}')
````
